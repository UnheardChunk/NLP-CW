```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Parsing the dataset

```python
dataset_path = "../dataset/original_datasets/dontpatronizeme_pcl.tsv"

dataset = pd.read_csv(dataset_path, sep="\t", skiprows=4, names=['par_id', 'art_id', 'keyword', 'country', 'text', 'orig_label'], index_col=0)
dataset.head()
```

```python
dataset.loc[dataset["text"].isna(), "text"] = ""
```

```python
dataset["label"] = dataset["orig_label"].apply(lambda x : 0 if (x == 0 or x == 1) else 1)
dataset.sample(10)
```

### Label frequency analysis

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
dataset["orig_label"].value_counts(sort=False).plot.bar(rot=0, ax=ax[0])
ax[0].set_xlabel("Original Label")
dataset["label"].value_counts(sort=False).plot.bar(rot=0, ax=ax[1])
ax[1].set_xlabel("New Label")
plt.plot()
```

```python
original_label_groups = dataset.groupby(["orig_label"])

original_label_groups["text"].agg(lambda x: x.str.len().mean())
```

### N-gram Frequency analysis

```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns

def get_top_n_grams(texts, n=None, ngram_range=(2,2)):
    """Gets the top n-grams with their frequencies"""
    vectorizer = CountVectorizer(ngram_range=ngram_range, stop_words="english")
    X = vectorizer.fit_transform(texts)
    frequencies = X.sum(axis=0).tolist()[0]
    ngrams_frequency = sorted(zip(vectorizer.get_feature_names_out(), frequencies), key=lambda x: x[1], reverse=True)[:n]

    return pd.DataFrame(ngrams_frequency, columns=["N-gram", "Frequency"])

def plot_ngrams(df, title, color):
    plt.figure(figsize=(10, 5))
    sns.barplot(x="Frequency", y="N-gram", data=df, palette=color)
    plt.title(title)
    plt.xlabel("Frequency")
    plt.ylabel("N-gram")
    plt.show()

# Get the bigrams for PCL and non-PCL texts
top_bigrams_pcl = get_top_n_grams(dataset[dataset['label'] == 1]["text"], n=10)
top_bigrams_nonpcl = get_top_n_grams(dataset[dataset['label'] == 0]["text"], n=10)
top_bigrams_pcl_orig = get_top_n_grams(dataset[dataset['orig_label'] == 4]["text"], n=10)
top_bigrams_nonpcl_orig = get_top_n_grams(dataset[dataset['orig_label'] == 0]["text"], n=10)

# Plot the bigrams
plot_ngrams(top_bigrams_pcl, "Top Bigrams in PCL Texts", "Blues_r")
plot_ngrams(top_bigrams_nonpcl, "Top Bigrams in Non-PCL Texts", "Greens_r")
plot_ngrams(top_bigrams_pcl_orig, "Top Bigrams in PCL Texts for label '4'", "Blues_r")
plot_ngrams(top_bigrams_nonpcl_orig, "Top Bigrams in Strictly Non-PCL Texts", "Greens_r")

```
